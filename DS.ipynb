```python
import numpy as np
import pandas as pd

np.random.seed(32)
data={"x": np.arange(1,10),
      "y": np.random.randint(15,100,size=9)
    }
df=pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (6,4))
plt.plot(df["x"],df["y"],marker='o',linestyle="--", color='blue')
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Line Plot")
plt.grid(True)
plt.show()
```
