```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv", encoding="latin1")
print("Unique customers:", df["CustomerID"].nunique())
```

```python
df["Revenue"] = df["Quantity"] * df["UnitPrice"]
invoice_revenue = df.groupby("InvoiceNo")["Revenue"].sum()
print(invoice_revenue.head())
```

```python
sns.histplot(df["UnitPrice"], bins=50)
plt.title("Distribution of Unit Price")
plt.show()
```

```python
top_countries = df["Country"].value_counts().head(5)

plt.figure(figsize=(6,4))
plt.bar(top_countries.index, top_countries.values)
plt.xlabel("Country")
plt.ylabel("Number of Purchases")
plt.title("Top 5 Purchasing Countries")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x=df["Quantity"])
plt.title("Quantity Outlier Detection")
plt.show()
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["Month"] = df["InvoiceDate"].dt.month

monthly_orders = df.groupby("Month")["InvoiceNo"].nunique()
print(monthly_orders)
```

```python
df["Revenue"] = df["Quantity"] * df["UnitPrice"]
monthly_revenue = df.groupby("Month")["Revenue"].sum()

plt.plot(monthly_revenue.index, monthly_revenue.values)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Monthly Revenue Trend")
plt.show()
```

```python
top4 = df["Country"].value_counts().head(4)

plt.pie(top4.values, labels=top4.index, autopct="%0.1f%%")
plt.title("Top 4 Countries by Purchase Share")
plt.show()
```
